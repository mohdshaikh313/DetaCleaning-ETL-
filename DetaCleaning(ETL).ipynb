```python
#Automating the process (optional)
def handle_missing_data(df, strategy='ffill'):
    if strategy == 'mean':
        return df.fillna(df.mean()) 
    elif strategy == 'median':
        return df.fillna(df.median())
    elif strategy == 'ffill':
        return df.fillna(df.ffill())
    else:
        return df.dropna()
```
